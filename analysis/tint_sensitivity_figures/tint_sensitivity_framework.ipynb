# Tint-sensitivity analysis framework
This notebook is a thin interactive wrapper around the tested `aurora_grid.tint.analysis` API. Change only the paths below; all validation, calculations, and exports remain in the shared Python package.

```python
from pathlib import Path
import pandas as pd
from aurora_grid.tint.analysis import generate_package, write_preflight_report

CONFIG = Path('params/tint_sensitivity_36.yaml')
INPUT_DIR = Path('/path/to/36_netcdfs')
PREFLIGHT_DIR = Path('results/tint_sensitivity_preflight')
FINAL_DIR = Path('results/tint_sensitivity_final')
```

## 1. Preflight
Run this before making figures. `final_ready` must be true before the final package is generated.

```python
write_preflight_report(CONFIG, INPUT_DIR, PREFLIGHT_DIR, overwrite=True)
inventory = pd.read_csv(PREFLIGHT_DIR / 'preflight_inventory.csv')
inventory
```

## 2. Generate the publication package
Use `mode='partial'` only for diagnostic work. Partial figures are visibly watermarked.

```python
generate_package(
    CONFIG,
    FINAL_DIR,
    mode='final',
    input_directory=INPUT_DIR,
    overwrite=True,
)
```

## 3. Inspect writeup tables

```python
abundances = pd.read_csv(FINAL_DIR / 'tables' / 'photospheric_abundances_1mbar.csv')
metrics = pd.read_csv(FINAL_DIR / 'tables' / 'sensitivity_metrics.csv')
display(abundances.head(), metrics.head())
```
